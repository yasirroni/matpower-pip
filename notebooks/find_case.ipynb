```python
import glob
import os

from matpowercaseframes import CaseFrames
from oct2py import Oct2PyError

from matpower import path_matpower, start_instance
```

```python
m = start_instance()
```

## Data location

Path to matpower installation can be found at `path_matpower`. The cases folder is
located at `f"{path_matpower}/data"` and `f"{path_matpower}/lib/t"` for standard cases
and test cases, respectively.

To find a case that has specific data, for example data with values of
`["RAMP_AGC", "RAMP_10", "RAMP_30"]` higher than `0`, we can loop read cases at
`path_matpower`.

```python
data_path = os.path.join(path_matpower, "data")

files_with_ramp_rates = []
ramp_cols = ["RAMP_AGC", "RAMP_10", "RAMP_30"]
for case_file in glob.glob(os.path.join(data_path, "case*.m")):
    try:
        try:
            cf = CaseFrames(case_file, load_case_engine=m)
        except Oct2PyError:
            print(
                f"Case {os.path.basename(case_file)} could not be loaded with Oct2Py "
                "and is loaded using parser instead."
            )
            cf = CaseFrames(case_file)
    except ValueError as e:
        print(f"Case {os.path.basename(case_file)} could not be loaded: {e}")
        continue
    existing_cols = [col for col in ramp_cols if col in cf.gen.columns]
    if (cf.gen[existing_cols] > 0).any().any():
        files_with_ramp_rates.append(os.path.basename(case_file))

files_with_ramp_rates
```

```python
# import start_instance to start matpower instance
from matpower import start_instance

# start instance
m = start_instance()
```

```python
cases = ['case9', 'case14', 'case30'] # list of test cases
for c in cases:
    mpc = m.eval(c, verbose=False) # load the case
    res = m.runpf(mpc) # run power flow
    print(c, res['success']) # print the result
```
